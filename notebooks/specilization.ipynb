```python
import pandas as pd
from collections import Counter, defaultdict

# Global variables
from pretty_html_table import build_table
```

```python
df_sdg = pd.read_pickle("../data/output/all_sdg_fixed_dst.pkl")
df_digital = pd.read_pickle("../data/dataframes/digital/all_digital.pkl")
```

```python
df_eu_countries = pd.read_excel("../data/countries_eu.xlsx", sheet_name="EU")
df_other = pd.read_excel("../data/countries_eu.xlsx", sheet_name="rest")
lst_eu = df_eu_countries.Country.to_list()
lst_other = df_other.Country.to_list()
```

```python
df_tot_pub = pd.read_excel("../data/total_number_pubs_2010_2022.xlsx", sheet_name=0, index_col=0)
dic_total_pubs = df_tot_pub.loc['TOT', :].to_dict()
```

```python
# list SDG
lst_sdg = ["SDG" + str(i) for i in range(1, 18)]
# list Categories
lst_cat = ["Environment", "Society", "Economy"]
# list DST
lst_dst = list(df_sdg.columns)[30:37]
lst_dst_shortened = ['AI', 'robotics', 'IOT']
```

## Categories

```python
d_cat = {country: defaultdict(lambda: 0) for country in lst_eu+lst_other}
d_dst = {country: defaultdict(lambda: 0) for country in lst_eu+lst_other}
d_cat_dst = {country: defaultdict(lambda: 0) for country in lst_eu+lst_other}
d_sdg = {country: defaultdict(lambda: 0) for country in lst_eu+lst_other}

# Categories + sdg
for ind, row in df_sdg.iterrows():
    lst_actors_in_row = row.CN.split(", ")
    for actor in lst_actors_in_row:
        if actor in d_cat.keys():
            for sdg in lst_sdg:
                if row[sdg]:
                    d_sdg[actor][sdg] += 1
            for cat in lst_cat:
                if row[cat]:
                    d_cat[actor][cat] += 1
# Digital
for ind, row in df_digital.iterrows():
    lst_actors_in_row = row.CN.split(", ")
    for actor in lst_actors_in_row:
        if actor in d_dst.keys():
            for dst in lst_dst:
                if row[dst]:
                    d_dst[actor][dst] += 1
# SDG-Digital
for ind, row in df_sdg[df_sdg['DST']].iterrows():
    lst_actors_in_row = row.CN.split(", ")
    for actor in lst_actors_in_row:
        if actor in d_cat_dst.keys():
            for dst in lst_dst_shortened:
                if row[dst]:
                    for cat in lst_cat:
                        if row[cat]:
                            d_cat_dst[actor][dst+"-"+cat] += 1
```

```python
lst_tableau = []
for elem in [d_cat, d_cat_dst, d_dst, d_sdg]:
    df_tableau = pd.DataFrame.from_dict(elem, orient='index')
    df_tableau.loc["total", :] = df_tableau.sum()
    df_tableau.loc[:, 'total'] = df_tableau.index.map(dic_total_pubs)
    df_tableau.iloc[-1, -1] = sum(dic_total_pubs[x] for x in df_tableau.iloc[0:-1].index)
    lst_tableau.append(df_tableau)
```

```python
df_tableau
```

```python
lst_df_spe = []
for df_tableau in lst_tableau:
    df_spe = df_tableau.copy(deep=True).iloc[0:-1, 0:-1]
    for col in df_spe.columns:
        for row in df_spe.index:
            df_spe.loc[row, col] = (df_tableau.loc[row, col] / df_tableau.loc[row, 'total']) / (df_tableau.loc["total", col] / df_tableau.loc["total", 'total'])
    lst_df_spe.append(df_spe)
```

```python
excel_path = "../img/Commission/specialization.xlsx"
with pd.ExcelWriter(excel_path) as writer:

    lst_df_spe[0].round(2).to_excel(writer, sheet_name='categories')
    lst_df_spe[1].round(2).to_excel(writer, sheet_name='cat-dt')
    lst_df_spe[2].round(2).to_excel(writer, sheet_name='dt')
    lst_df_spe[3].round(2).to_excel(writer, sheet_name='sdg')
```

##  Normalising by pop

```python
eu_countries = pd.read_excel("../data/countries_eu.xlsx", sheet_name='EU').Country.to_list()
other_countries = pd.read_excel("../data/countries_eu.xlsx", sheet_name='rest').Country.to_list()
gdp = pd.read_csv("../data/countriesgdp_pop.csv", index_col="ranking")
gdp = gdp[gdp.country.isin(eu_countries + other_countries)].set_index("country")
gdp.loc["EU", "gdppc"] = 44024
gdp.loc["EU", "pop2022"] = 446800
gdp
```

```python
lst_tableau_normed = []

for tab in lst_tableau:
    tab_normed = tab.copy(deep=True).iloc[0:-1,0:-1]
    tab_normed.loc[:,"pop"] = tab_normed.index.map(gdp.pop2022.to_dict())
    copy_columns = list(tab_normed.columns)
    copy_columns.remove('pop')
    for col in copy_columns:
        tab_normed.loc[:,col] = tab_normed.apply(lambda x:x[col]/x['pop'], axis=1)
    tab_normed.drop(columns=['pop'], axis=1, inplace=True)
    tab_normed.loc[:,'total_per_country'] = tab_normed.loc[:,:].sum(axis=1)
    tab_normed.loc["total_per_cat", :] = tab_normed.loc[:,:].sum(axis=0)

    lst_tableau_normed.append(tab_normed)
```

```python
lst_df_spe = []
for df_tableau in lst_tableau_normed:
    df_spe = df_tableau.copy(deep=True).iloc[0:-1, 0:-1]
    for col in df_spe.columns:
        for row in df_spe.index:
            df_spe.loc[row, col] = (df_tableau.loc[row, col] / df_tableau.loc[row, 'total']) / (
                        df_tableau.loc["total", col] / df_tableau.loc["total", 'total'])
    lst_df_spe.append(df_spe)
```

```python
excel_path = "../img/Commission/specialization.xlsx"
with pd.ExcelWriter(excel_path) as writer:

    lst_df_spe[0].round(2).to_excel(writer, sheet_name='categories_norm_pop')
    lst_df_spe[1].round(2).to_excel(writer, sheet_name='cat-dt')
    lst_df_spe[2].round(2).to_excel(writer, sheet_name='dt_norm_pop')
    lst_df_spe[3].round(2).to_excel(writer, sheet_name='sdg_norm_pop')
```
